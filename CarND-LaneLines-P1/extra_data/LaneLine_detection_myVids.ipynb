```python
def make_snapshots(num_frames, vid, img_dir):
    """
    This function takes a video and generates several images/snapshots 
    of the videos at different time.
    
    num_frames - number of images to generate
    vid - video clip
    img_dir - directory where to save the images
    """
    vid_length = vid.duration
    for i in range(num_frames):
        time_step = vid_length%num_frames
        clip.save_frame(fname +'_'+str(i)+'.png', t=i, withmask=False)
        img = Image.open(fname +'_'+str(i)+'.png')
        img.save(img_dir+'/'+fname +'_'+str(i)+'.jpeg', "JPEG")
        

# Video Processing

def create_ShortVideo(init_vid, start_clip, end_clip):
    """
    Take a shorter section of the video
    """
    duration = clip.duration
    if end_clip > duration:
        end_clip = duration
        
    try:
        return init_vid.subclip(start_clip, end_clip)
    except ValueError:
        print('problem creating the clip')
    
```

```python
################
# PARAMETERS
################
fname = 'vid10.mov'
start_clip, end_clip = [1, 45]
#crop from top y1 and crop y_end from bottom
crop_window = {'y1': 0, 'height':700}
set_clip_height = 420 #height of movie
num_frames = 10

clip = VideoFileClip(fname, audio=False)

# Get a short Video
clip = create_ShortVideo(clip, start_clip, end_clip) 
#Crop a rectangle that starts 10 pixels left and is 200px wide
clip = clip.crop(y1=crop_window['y1'] , height=crop_window['height'])
#resize video
clip = clip.resize(height=set_clip_height)

#save a few frames
img_dir = fname[0:-4]+'_imgs'

if not os.path.exists(img_dir):
    os.makedirs(img_dir)
    make_snapshots(num_frames, clip, img_dir)
else:
    make_snapshots(num_frames, clip, img_dir)
    
clip.write_videofile(fname +'_edit.mp4',audio=False) # don't render audio.
```

```python
##########
# List of transformations
##########

import math


def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def hough(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
    minLineLength - Minimum length of line. Line segments shorter than this are rejected.
    maxLineGap - Maximum allowed gap between line segments to treat them as single line.
    Returns the two endpoints of lines.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    return lines
            
def wavg_lines(img, lines, angle_threshold, y_min, y_max, color=[0, 255, 0], thickness=2):
    """
     Return coordinates (x1,y1,x2,y2) of the weighted average of the Hough Lines
     The weighted average takes into account the norm of the segments as the weighting parameter.
     The idea is that the extended line shoudl follow preferentially the long segments that are likely associated
     to real lane line rather than the short lines.
    """
    hough_pts = {'m_left': [], 'b_left':[], 'norm_left':[], 'm_right': [], 'b_right':[], 'norm_right':[]}
    if lines != None : #if no lines left or right are detected
        for line in lines:
            for x1,y1,x2,y2 in line:
                angle = abs(np.arctan( (y2-y1)/(x2-x1) ) * 180/np.pi )
                if angle > angle_threshold:
                    m, b = np.polyfit([x1,x2], (y1,y2), 1)
                    norm = ((x2-x1)**2 + (y2-y1)**2)**0.5
                    #print("angle_good")
                    if m > 0: #right
                        #print("positive m")
                        hough_pts['m_right'].append(m*norm)
                        hough_pts['b_right'].append(b*norm)
                        hough_pts['norm_right'].append(norm)

                    if m < 0: #left                    
                        #print("negative m")
                        hough_pts['m_left'].append(m*norm)
                        hough_pts['b_left'].append(b*norm)
                        hough_pts['norm_left'].append(norm)
    
    #for each lane line: calculate weighted average of slope (m) and intercept (b)
    #calculate weighted average if the lists are not empty
    if len(hough_pts['b_left']) != 0 or len(hough_pts['m_left']) != 0 or len(hough_pts['norm_left']) != 0:
        b_avg_left = np.sum( np.array(hough_pts['b_left']) )/np.sum( np.array(hough_pts['norm_left'] ) )
        m_avg_left = np.sum( np.array(hough_pts['m_left']) )/np.sum( np.array(hough_pts['norm_left'] ) )
        xmin_left = int((y_min - b_avg_left)/m_avg_left)
        xmax_left = int((y_max - b_avg_left)/m_avg_left)
        left_lane = [[xmin_left, y_min, xmax_left, y_max]]
    else:
        left_lane = [[0, 0, 0, 0]]

    if len(hough_pts['b_right']) != 0 or len(hough_pts['m_right']) != 0 or len(hough_pts['norm_right']) != 0:
        b_avg_right = np.sum( np.array(hough_pts['b_right']) )/np.sum( np.array(hough_pts['norm_right'] ) )
        m_avg_right = np.sum( np.array(hough_pts['m_right']) )/np.sum( np.array(hough_pts['norm_right'] ) )
        xmin_right = int((y_min - b_avg_right)/m_avg_right)
        xmax_right = int((y_max - b_avg_right)/m_avg_right)
        right_lane =  [[xmin_right, y_min, xmax_right, y_max]]
    else:
        right_lane =  [[0, 0, 0, 0]]

    return [left_lane, right_lane]

            
def avg_lines(img, lines, angle_threshold, y_min, y_max, color=[0, 0, 255], thickness=2):
    """
     return coordinates (x1,y1,x2,y2) of the regular average of the Hough Lines
    """
    hough_pts = {'m_left': [], 'b_left':[], 'norm_left':[], 'm_right': [], 'b_right':[], 'norm_right':[]}
    if lines != None : #if no lines left or right are detected
        for line in lines:
            for x1,y1,x2,y2 in line:
                angle = abs(np.arctan( (y2-y1)/(x2-x1) ) * 180/np.pi )
                if angle > angle_threshold:
                    m, b = np.polyfit([x1,x2], (y1,y2), 1)
                    norm = ((x2-x1)**2 + (y2-y1)**2)**0.5
                    #print("angle_good")
                    if m > 0: #right
                        #print("positive m")
                        hough_pts['m_right'].append(m)
                        hough_pts['b_right'].append(b)
                        hough_pts['norm_right'].append(norm)

                    if m < 0: #left                    
                        #print("negative m")
                        hough_pts['m_left'].append(m)
                        hough_pts['b_left'].append(b)
                        hough_pts['norm_left'].append(norm)
    
    #for each lane line: calculate weighted average of slope (m) and intercept (b)
    #calculate weighted average if the lists are not empty
    if len(hough_pts['b_left']) != 0 or len(hough_pts['m_left']) != 0 or len(hough_pts['norm_left']) != 0:
        b_avg_left = np.mean( np.array(hough_pts['b_left']) )
        m_avg_left = np.mean( np.array(hough_pts['m_left']) )
        xmin_left = int((y_min - b_avg_left)/m_avg_left)
        xmax_left = int((y_max - b_avg_left)/m_avg_left)
        left_lane = [[xmin_left, y_min, xmax_left, y_max]]
    else:
        left_lane = [[0, 0, 0, 0]]
        
    if len(hough_pts['b_right']) != 0 or len(hough_pts['m_right']) != 0 or len(hough_pts['norm_right']) != 0:
        b_avg_right = np.mean( np.array(hough_pts['b_right']) )
        m_avg_right = np.mean( np.array(hough_pts['m_right']) )
        xmin_right = int((y_min - b_avg_right)/m_avg_right)
        xmax_right = int((y_max - b_avg_right)/m_avg_right)
        right_lane = [[xmin_right, y_min, xmax_right, y_max]]
    else:
        right_lane = [[0, 0, 0, 0]]
        
    return [left_lane, right_lane]
    
    
    
def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    if lines != None : #if no lines left or right are detected
        for line in lines:
            for x1,y1,x2,y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)

# Python 3 has support for cool math symbols.
def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.   
    `initial_img` should be the image before any processing.
    The result image is computed as follows:
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

```python
import os
dir_img = "vid17_imgs/"
f = os.listdir(dir_img)
f = [fname for fname in f if 'jpeg' in fname or 'jpg']


'''
display_steps = all (show images at all steps of pipeline) / ='final' (show final image )
hline_show - 'avg' / 'weighted
hline_show = {'hough_lines: on/off, 'avg': on/off, 'w_avg': on/off}
'''

def pipeline(img, vertices, threshold_angle, hline_show ):
    
    #convert to grayscale
    gray = grayscale(img)
    
    #Apply Gaussian Blur: noise smooth noise
    gray_blur = gaussian_blur(gray, 3)
    
    #apply Canny Edge Detector
    edges = canny(gray_blur, 10, 180)

    #apply mask
    imshape = img.shape
    masked = region_of_interest(edges, vertices)
    hough_lines = hough(masked, rho=1, theta=np.pi/180, threshold=25, min_line_len=10, max_line_gap=10)
    
    if hline_show['hlines'] == 'on':                               
        hlines_img = np.zeros(imshape, dtype=np.uint8)
        draw_lines(hlines_img, hough_lines, color=[255, 0, 0], thickness=2)
    else:
        hlines_img = np.zeros(imshape, dtype=np.uint8)    
    
    if hline_show['w_avg'] == 'on':                           
        w_avg_hlines = wavg_lines(img, hough_lines, threshold_angle, int(img.shape[0]*0.85), \
                                  img.shape[0], color=[255, 0, 0], thickness=2)
        w_avg_img = np.zeros(imshape, dtype=np.uint8)
        draw_lines(w_avg_img, w_avg_hlines, color=[0, 255, 0], thickness=10)

    else:
        w_avg_img = np.zeros(imshape, dtype=np.uint8)
        
    if hline_show['avg'] == 'on':                           
        avg_hlines = avg_lines(img, hough_lines, threshold_angle, int(img.shape[0]*0.85), img.shape[0], \
                               color=[255, 0, 0], thickness=2)
        avg_img = np.zeros(imshape, dtype=np.uint8)
        draw_lines(avg_img, avg_hlines, color=[0, 0, 255], thickness=10)

    else:
        avg_img = np.zeros(imshape, dtype=np.uint8)    
    
    img_all_lines = weighted_img(hlines_img, img, α=1, β=0.5, λ=0.)
    img_all_lines = weighted_img(w_avg_img, img_all_lines, α=1, β=0.5, λ=0.)
    img_all_lines = weighted_img(avg_img, img_all_lines, α=1, β=0.5, λ=0.)
    return img_all_lines




for img_name in f:
    """
    Apply the lane line detection/tracking pipeline to a few images 
    """
    #reading in an imag
    print('Image:', img_name)
    img = mpimg.imread(dir_img+img_name)
    hline_show = {'hlines': 'on', 'w_avg': 'on', 'avg': 'off'}
    imshape = img.shape
    vertices = np.array([[(imshape[1]*0.05,imshape[0]), (140,imshape[0]*0.7), \
                          (940, imshape[0]*0.7), (imshape[1],imshape[0]), \
                          (imshape[1]*0.05,imshape[0])]], dtype=np.int32)
    threshold_angle = 25
    lines_img = pipeline(img, vertices, threshold_angle, hline_show)
    plt.imshow(lines_img)
    plt.show()


```

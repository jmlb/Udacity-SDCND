```python
def camera_calibration(folder, nx, ny, showMe=0):
    '''
    find all the location of all corners (x, y, z) with z=0
    using openCV findChessBoardCorners
    return a dictionary:
        ret: True/False if corners were detected by openCV findChessboardCorners
        imagePoints: coordinate of corners detected in all images
        objectPoints: coordinate of corners in destination image (x, y, z)
        img_sz: size of image (width, height)
        ''
        run calibration using corners[ret, imgPoints, objPoints, img_sz] found in all images
        return:
            ret: RMSError of calibration
            mtx: Camera Matrix
            dist: distorsion coefficient
            rotationVec: rotation vecteurs
            tvecs: translation vectors
    '''
    # Store object points and image points from all the images
    objpoints = [] #3D points in real world space
    imgpoints = [] #2D points in image plane
    #Prepare object points, like (0,0,0), (1, 0,0)
    objp = np.zeros((nx * ny, 3), np.float32)
    objp[:,:2] = np.mgrid[0:nx,0:ny].T.reshape(-1,2) #x, y coordinate
        
    assert len(folder) != 0, 'No file found in folder'
        
    for fname in folder:
        img = cv2.imread(fname)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        ret, corners = cv2.findChessboardCorners(gray, (nx, ny), None)
        img_sz = gray.shape[::-1]
            
        if ret:
            imgpoints.append(corners)
            objpoints.append(objp)
            if showMe:
                draw_corners = cv2.drawChessboardCorners(img, (nx, ny), corners, ret)
                display(draw_corners, 'Found all corners: {} '.format(ret))

    if len(objpoints) == len(imgpoints) and len(objpoints) != 0:
        print('ok')
        ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_sz, None, None)
        return {'ret': ret, 'cameraMatrix': mtx, 'distorsionCoeff': dist, \
                    'rotationVec': rvecs, 'translationVec': tvecs}
    else:
        raise Error('Camera Calibration failed')


# Distorsion correction        
def correction(image, calib_params, showMe=0):
    corrected = cv2.undistort(image, calib_params['cameraMatrix'], calib_params['distorsionCoeff'], None, 
                         calib_params['cameraMatrix'])
    if showMe:
        display(image, 'Original', color=1)
        display(corrected, 'After correction', color=1)

        

nx = 9 #number of corners in a row
ny = 6 #numbers or corners in a column
folder_calibration = glob.glob("camera_cal/calibration*.jpg") #list of chessboard image file
calib_params = camera_calibration(folder_calibration, nx, ny, showMe=0)
# for a good calibration ret must be between 0.1 and 1.0
print('RMS Error of Camera calibration: {:.3f}'.format(calib_params['ret']) )
print('This number must be between 0.1 and 1.0')

#Show example distorsion correction
imgs_tests = glob.glob("test_images/*.jpg")
original_img = np.random.choice(imgs_tests)
original_img = cv2.imread("test_images/test5.jpg")
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
corr_img = correction(original_img, calib_params, showMe=showMe)
```

# Image process

```python
class ImageProcess():
    '''
    Processing methods of original images
    '''
    def __init__(self):
        pass
    
    def gaussianBlur(self, img, k_sz):
        k_sz = 5
        img = cv2.GaussianBlur(img, (k_sz, k_sz), 0)
        return img
    
    
    def directional_gradient(self, img, direction='x', thresh=[0, 255]):
        '''
        Gradient along vertical or horizontal direction using OpenCV Sobel 
        img: RGB or Grayscale image
        direction: x(horizontal) or y(vertical) for gradient direction
        thresh: apply threshold on pixel intensity of gradient image
        ksize: kernel size (can only be a odd number)
        output is a binary image
        '''
        if direction == 'x':
            sobel = cv2.Sobel(img, cv2.CV_64F, 1, 0)
        elif direction == 'y':
            sobel = cv2.Sobel(img, cv2.CV_64F, 0, 1)
        
        sobel_abs = np.absolute(sobel)  #absolute value
        scaled_sobel = np.uint8(sobel_abs * 255/np.max(sobel_abs)) #turn sobel to 8bit image 0-255 intensity range
        binary_output = np.zeros_like(sobel)
        binary_output[(scaled_sobel>= thresh[0]) & (scaled_sobel <= thresh[1]) ] = 1 #generate binary
        
        return binary_output


    def mag_gradient(self, img, thresh=[0, 255]):
        '''
        Magnitude of gradient : sqrt(gradx**2 + grady**2)
        img: RGB or Grayscale image
        thresh: apply threshold on pixel intensity of the gardient magnitude
        ksize: kernel size (can only be a odd number)
        output is a binary image
        '''
        sobelx = cv2.Sobel(img, cv2.CV_64F, 1, 0)  #gradient along x
        sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1) #gradient along y
        gradient_mag = np.sqrt( np.square(sobelx) + np.square(sobely)) # norm of gradient
        scaled_gradient_mag = np.uint8(gradient_mag * 255/np.max(gradient_mag)) #turn sobel to 8bit image 0-255 intensity range
        binary_output = np.zeros_like(gradient_mag)
        binary_output[(scaled_gradient_mag >= thresh[0]) & (scaled_gradient_mag <= thresh[1]) ] = 1 #thresholding
        
        return binary_output


    def gradient_direction(self, img, thresh=[0, 90], ksize=3):
        '''
        Direction of gradient: arctan(grady/gradx)
        img: RGB or Grayscale image
        thresh: apply threshold on gradient direction in degrees (0, 90)
        ksize: kernel size (can only be a odd number)
        output is a binary image
        '''
        sobelx = cv2.Sobel(img, cv2.CV_64F, 1,0, ksize) #gradient along x
        sobely = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize) #gradient along y
        gradient_dir = np.arctan2( sobely, sobelx)
        thresh = [thresh[0] * np.pi/180, thresh[1] * np.pi/180] #convert threshold from degree to radian
        binary_output = np.zeros_like(gradient_dir)
        binary_output[(gradient_dir>= thresh[0]) & (gradient_dir <= thresh[1]) ] = 1 #thresholding
        
        return binary_output

    
    def color_binary(self, img, dst_format='HLS', ch=2, ch_thresh=[0,255]):
        '''
        Color thesholding on channel 2 and 3
        img: RGB
        dst_format: destination format (HLS or HSV)
        ch2_thresh: pixel intensity threshold on channel 2
        ch3_thresh: pixel intensity threshold on channel 3
        output is a binary image
        '''
        if dst_format == 'HSV':
            img = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
            ch_binary = np.zeros_like(img[:,:, int(ch-1)])
        else:
            img = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
            ch_binary = np.zeros_like(img[:,:, int(ch-1)])
            ch_binary[(img[:,:,int(ch-1)] >= ch_thresh[0]) & (img[:,:,int(ch-1)]<= ch_thresh[1])] = 1
        
        return ch_binary
    
    
    
    def image_correction(self, img, cal_params):
        '''
        correct image from camera distorsion
        img: original image RGB format
        cal_params: calibration parameters of camera
        return: undistorted image
        '''
        dst = cv2.undistort(img, cal_params['cameraMatrix'], \
                            cal_params['distorsionCoeff'], None, \
                            cal_params['cameraMatrix'])
        return dst
        
    
    def convert2_rgb(self, img):
        '''
        convert image to RGB
        img: RGB image
        '''
        try:
            rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            return rgb
        except:
            print('image cannpot be converted to RGB')
    
    
    def convert2_gray(self, img):
        '''
        convert image to gray
        img: RGB image
        '''
        if len(img.shape) == 3:
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            return gray
        elif len(img.shape) == 2: #img channel already squashed
            return img

    
    def birdView(self, img, birdview_pts):
        '''
        Transform image to birdview
        birdview_pts: dict with 
                    * src_pts: source points (points in original image)
                    * dst_pts: destination points (points in birdview image)
        '''
        M = cv2.getPerspectiveTransform(birdview_pts['src_pts'], birdview_pts['dst_pts'])
        Minv = cv2.getPerspectiveTransform(birdview_pts['dst_pts'], birdview_pts['src_pts'])
        img_sz = (img.shape[1], img.shape[0])
        img_warped = cv2.warpPerspective(img, M, img_sz, flags=cv2.INTER_LINEAR)
        
        return (img_warped, Minv)
    

    
# Image processing pipeline

def image_process_pipeline(image, camera_calib, k_blur, birdview_pts, gradx_thresh=[0, 255], grady_thresh=[0, 255], 
                     ch_thresh=[0, 255], showMe=0):
    '''
    Image processing to highlight lanes
    image: image BGR format (OpenCV)
    camera_calib: calibration parameters of camera
    birdview_pts: source and destination points
    showMe: if True show images at each process step
    '''

    process = ImageProcess()
    
    image = process.image_correction(image, camera_calib)
    if showMe: display(image, 'Apply Camera Correction', color=1)
        
    #Gaussian Blur: Noise removal
    #image = process.gaussianBlur(image, k_blur)
    #if showMe: display(image, 'Gaussian Blur', color=1)
    
    gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    if showMe: display(gray, 'GRAY', color=0)
        
    gradx = process.directional_gradient(gray, direction='x', thresh=gradx_thresh )
    if showMe: display(gradx, 'Gradient x', color=0)
    
    ch3_hls_binary = process.color_binary(image, dst_format='HLS', ch=3, ch_thresh=ch_thresh)
    if showMe: display(ch3_hls_binary, 'HLS to Binary S ', color=0)
        
    ch3_hls_grad = process.directional_gradient(ch3_hls_binary, direction='x', thresh=gradx_thresh )
    if showMe: display(gradx, 'Gradient x HLS', color=0)
    
    combined_gradx = np.zeros_like(gradx)
    combined_gradx[((gradx == 1) | (ch3_hls_grad == 1) )] = 1
    if showMe: display(combined_gradx, 'Combined Gradient x of HLS and Gray', color=0)
        
    combined_output = np.zeros_like(gradx)
    combined_output[((combined_gradx == 1) | (ch3_hls_binary == 1) )] = 1
    if showMe: display(combined_output, 'Combined output', color=0)
    
    imshape = combined_output.shape
    mask = np.zeros_like(combined_output)
    vertices = np.array([[(100, 710), (520, 480), (790, 480), (1210, 710)]], dtype=np.int32)
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, 1)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(combined_output, mask)
    
    cleaned = morphology.remove_small_objects(masked_image.astype('bool'), min_size=10, connectivity=2)
    if showMe: display(cleaned, 'cleaned', color=0)
    
    output_warped, Minv = process.birdView(cleaned*1.0, birdview_pts)
    if showMe: display(output_warped, 'Bird View', color=0)
        
    return (output_warped, Minv)
```

# Finding Lane

```python
import matplotlib.patches as patches


class FindingLanes():
    '''
    extract pixels associated with lanes and fit them to 2nd order polynomial function
    '''

    
    def __init__(self, image, bottom_crop, sliding_window, hot_pixel_thresh):
        self.image = image
        self.bottom_crop = bottom_crop
        self.hot_pixel_thresh = hot_pixel_thresh
        self.sliding_window = sliding_window
        #adjust image height (remove front-end of car)
        self.image = self.image[0:self.bottom_crop, :]
 
    
    def MahalanobisDist(self, x, y):
        covariance_xy = np.cov(x,y, rowvar=0)
        inv_covariance_xy = np.linalg.inv(covariance_xy)
        xy_mean = np.mean(x),np.mean(y)
        x_diff = np.array([x_i - xy_mean[0] for x_i in x])
        y_diff = np.array([y_i - xy_mean[1] for y_i in y])
        diff_xy = np.transpose([x_diff, y_diff])
    
        md = []
        for i in range(len(diff_xy)):
            md.append(np.sqrt(np.dot(np.dot(np.transpose(diff_xy[i]),inv_covariance_xy),diff_xy[i])))
        return md


    def MD_removeOutliers(self, x, y):
        MD = self.MahalanobisDist(x, y)
        threshold = np.mean(MD) * 1.9 # adjust 1.5 accordingly 
        nx, ny, outliers = [], [], []
        for i in range(len(MD)):
            if MD[i] <= threshold:
                nx.append(x[i])
                ny.append(y[i])
            else:
                outliers.append(i) # position of removed pair
        return (np.array(nx), np.array(ny), np.array(outliers))

    
    def polynomial_fit(self, data):
        '''
        Perform 2nd order polynomial fit: a0 + a1 x + a2 x**2
        data: dictionary with x and y values {'var':[], 'output':[]}
        '''
        #nx, ny, _ = MD_removeOutliers(data['var'], data['output'])
        a2, a1, a0 = np.polyfit(data['var'], data['output'], 2)
        return {'a0': a0, 'a1': a1, 'a2': a2}


        
    def find_starter_centroids(self, hot_pixel_thresh):
        '''
        Find starter centroids (for left and right lane) using histogram of bottom half of the image 
        hot_pixel_thresh: If peak intensity is below a threshold use histogram on the full height of the image
        returns left and right centroid position
        '''
        window_l = {'x0': 0, 'y0': self.image.shape[0], 
                    'width':self.image.shape[1]/2, 'height':self.image.shape[0]/2}
        centroid_l, hot_pixel_cnt_l = self.find_centroid(window_l)
        if (hot_pixel_cnt_l < hot_pixel_thresh):
            window_l['height']=self.image.shape[0]
            centroid_l, hot_pixel_cnt_l = self.find_centroid(window_l)
            
        window_r = {'x0': self.image.shape[1]/2, 'y0': self.image.shape[0], 'width':self.image.shape[1]/2, 
                    'height':self.image.shape[0]/2}
        centroid_r, hot_pixel_cnt_r = self.find_centroid(window_r)
        if (hot_pixel_cnt_r < hot_pixel_thresh):   
            window_r['height']=self.image.shape[0]
            centroid_r, hot_pixel_cnt_r = self.find_centroid(window_r)
        #only interested in rough estimate of the line
        return {'left': [centroid_l, hot_pixel_cnt_l], 'right':[centroid_r, hot_pixel_cnt_r]}

    
    
    def find_centroid(self, window, showMe=showMe):
        '''
        find centroid in window using histogram of pixels within the window
        img: binary image
        window: dict({'x0', 'y0', 'width', 'height'})
        display if hotpixel_cnt (number of pixel in window is smaller than a threshold, 
        take centroid as center of window)
        '''
        #crop image to window dimension
        crop_image = self.image[ round( window['y0'] - window['height']):round(window['y0']), 
                              round(window['x0']):round(window['x0']+ window['width'])]
        histogram = np.sum(crop_image, axis=0)
        if showMe:
            plt.plot(histogram)
            plt.title('Histogram')
            plt.show()
        hot_pixel_cnt = np.sum(histogram)
        if hot_pixel_cnt > self.hot_pixel_thresh:
            try:
                centroid = np.argmax(histogram)
                cnt = histogram[centroid]
            except:
                print('Centroid not found!')
            centroid = int( round(centroid + window['x0']) )
        
        else:
            #take center of box as new centroid ()
            centroid = int( round(window['x0'] + window['width']/2) )
            cnt=0
            
        print(centroid)
        return (centroid, cnt)   
    
 
    def run_sliding_window(self, centroid_starter_l, centroid_starter_r, showMe=showMe):
        '''
        Run sliding window from bottom of image to top and return indexes of the hot pixels associated with lane
        centroi_starter_l: centroid starting location for left lane
        centroid_starter_r: centroid starting location for right lane
        '''
        #Initialize window for left and right lane
        window_l = {'x0': centroid_starter_l - int(self.sliding_window['width']/2), 'y0': self.image.shape[0],
                 'width': self.sliding_window['width'], 
                  'height': round(self.image.shape[0]/self.sliding_window['n_steps'])}

        window_r = {'x0': centroid_starter_r - int(self.sliding_window['width']/2), 'y0': self.image.shape[0],
                 'width': self.sliding_window['width'], 
                  'height': round(self.image.shape[0]/self.sliding_window['n_steps'])}
        
        #Initialize dict to store (var_height, output=f(var)) coordinate of hotpixels and centroid location after all sliding window steps
        #at each step of sliding window
        #
        hotpixels_log = { 'left': {'var': [], 'output':[]}, 'right': {'var':[], 'output':[]} }
        centroids_log = {'left': [], 'right': []}
        if showMe:
            out_img = (np.dstack((self.image, self.image, self.image) ) * 255).astype('uint8') #convert to uint8
        
        for step in range(self.sliding_window['n_steps']):
            
            #Limit lateral position of box, must remain within image width
            if (window_l['x0'] < 0): window_l['x0'] = 0   
            if (window_r['x0'] + window_r['width']) > self.image.shape[1]: 
                window_r['x0'] = self.image.shape[1] - window_r['width']
            
            centroid_r, hot_pixel_cnt_r = self.find_centroid(window_r)
            centroid_l, hot_pixel_cnt_l = self.find_centroid(window_l)
            
            # Create a copy of image where all pixels outside right/left lane window are turn off (=0)
            masked_l = np.zeros_like(self.image)
            masked_l[ window_l['y0']- window_l['height'] : window_l['y0'],
                                window_l['x0']:window_l['x0']+window_l['width']] \
                    = self.image[ window_l['y0']- window_l['height'] : window_l['y0'],
                                window_l['x0']:window_l['x0']+window_l['width']]
            
            masked_r = np.zeros_like(self.image)
            masked_r[ window_r['y0']- window_r['height'] : window_r['y0'],
                                window_r['x0']:window_r['x0']+window_r['width']] \
                    = self.image[window_r['y0']- window_r['height'] : window_r['y0'],
                                window_r['x0']:window_r['x0']+window_r['width']]
                
            
            #Get coordinates of hot pixels in left/right window
            hot_pixels_l = np.nonzero( masked_l )
            hot_pixels_r = np.nonzero( masked_r )
            
            #Do not record pixels if total count of hotpixels below thresh
            if len(hot_pixels_l[0]) >= self.hot_pixel_thresh:
                hotpixels_log['left']['var'].extend(hot_pixels_l[0].tolist())
                hotpixels_log['left']['output'].extend(hot_pixels_l[1].tolist())
            
            if len(hot_pixels_r[0]) >= self.hot_pixel_thresh:
                hotpixels_log['right']['var'].extend(hot_pixels_r[0].tolist())
                hotpixels_log['right']['output'].extend(hot_pixels_r[1].tolist()) 
           
            #Draw windows
            if showMe:
                cv2.rectangle(out_img, 
                          (window_l['x0'], window_l['y0']- window_l['height']),
                          (window_l['x0']+window_l['width'], window_l['y0']),(0,255,0), 2)
                cv2.rectangle(out_img, 
                          (window_r['x0'], window_r['y0']- window_r['height']),
                          (window_r['x0']+window_r['width'], window_r['y0']),(0,0,255), 2)
                plt.imshow(out_img)
                plt.show()
            #update record of centroids
            centroids_log['left'].append([centroid_l, window_l['y0']])
            centroids_log['right'].append([centroid_r, window_r['y0']])
            #window next position
            window_l['x0'] = round(centroid_l - window_l['width']/2)
            window_l['y0'] = window_l['y0'] - window_l['height']
            window_r['x0'] = round(centroid_r - window_r['width']/2)
            window_r['y0'] = window_r['y0'] - window_r['height']
        np.save('r_output.npy', hotpixels_log['right']['output'])
        np.save('r_var.npy', hotpixels_log['right']['var'])
        np.save('l_output.npy', hotpixels_log['left']['output'])
        np.save('l_var.npy', hotpixels_log['left']['var'])
        return hotpixels_log
```

# Example of image thru imageProcessing

```python
k_blur = 5
src_pts = np.float32([[200, 710], [520, 500], [770, 500], [1110, 710]])
dst_pts = np.float32([[200, 720], [200, 20], [1010, 20], [1000, 720]])
showMe = 0
bottom_crop = -60 #front-end car
sliding_window_specs = {'width': 160, 'n_steps': 9} #number of steps vertical steps of sliding window
hot_pixel_thresh = 50 # if number of hot pixel in window below 50, #consider them as noise and do not attempt to get centroid
    

ym_per_pix = 30/720 # meters per pixel in y dimension
xm_per_pix = 3.7/700 # meters per pixel in x dimension
birdview_pts = {'src_pts': src_pts, 'dst_pts': dst_pts}
#input_image = cv2.imread('test_images/straight_lines1.jpg')


draw_lane_thickness = 20 #hwo thick to draw predicted lanes in px
smoothing = 9


# Define a class to receive the characteristics of each line detection
class Line():
    def __init__(self, buffer_sz):
        # was the line detected in the last iteration?
        self.detected = False  
        # x values of the last n fits of the line
        self.recent_xfitted = [] 
        #average x values of the fitted line over the last n iterations
        self.bestx = None     
        #polynomial coefficients averaged over the last n iterations
        self.best_fit = None  
        #polynomial coefficients for the most recent fit
        self.current_fit = [np.array([False])]  
        #radius of curvature of the line in some units
        self.radius_of_curvature = None 
        #distance in meters of vehicle center from the line
        self.line_base_pos = None 
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float') 
        #x values for detected line pixels
        self.allx = None  
        #y values for detected line pixels
        self.ally = None
        self.buffer_sz = buffer_sz
        
        


            

def pipeline_predict_lanes(input_image, showMe=showMe):
    
    k_blur = 5
    sliding_window_specs = {'width': 80, 'n_steps': 9}
    bottom_crop = -60 #front-end car
    sliding_window_specs = {'width': 160, 'n_steps': 9} #number of steps vertical steps of sliding window
    hot_pixel_thresh = 100 # if number of hot pixel in window below 50, #consider them as noise and do not attempt to get centroid
    
    img_sz = (input_image.shape[1], input_image.shape[0])
    ym_per_pix = 30/720 # meters per pixel in y dimension
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    #birdview_pts = {'src_pts': src_pts, 'dst_pts': dst_pts}
    #input_image = cv2.imread('test_images/straight_lines1.jpg')
    ref_point_radius = input_image.shape[0]

    draw_lane_thickness = 20 #hwo thick to draw predicted lanes in px
    #input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
    # Run image processing pipeline
    input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
    warped_img, Minv = image_process_pipeline(input_image, calib_params, k_blur,  birdview_pts, 
                                gradx_thresh=[25, 255], 
                                ch_thresh=[70, 255],
                                showMe=showMe)
    #print('Values:.....', np.sum(warped_img))
    
    w_img = np.zeros_like(warped_img)
    w_img[warped_img >= 0.5] =1
    warped_img = w_img
    #Run lane finding
    lanes = FindingLanes(warped_img, bottom_crop, sliding_window_specs, hot_pixel_thresh)
    #hot_pixel_thresh : if the half window contain hot pixel below the threshold, take histogram on full height
    centroid_starters = lanes.find_starter_centroids(hot_pixel_thresh=50)
    centroid_starter_right = centroid_starters['right']
    centroid_starter_left = centroid_starters['left']
    # record left/right lane hotpixels coordinates x and y
    hotpixels_log = lanes.run_sliding_window(centroid_starter_left[0], centroid_starter_right[0])
    #print( hotpixels_log['right'] )
    
    # check out outliers
    
    hotpixels_log['left']['var'], hotpixels_log['left']['output'], _ = \
                    lanes.MD_removeOutliers(hotpixels_log['left']['var'], hotpixels_log['left']['output'])

    hotpixels_log['right']['var'], hotpixels_log['right']['output'], _ = \
                    lanes.MD_removeOutliers(hotpixels_log['right']['var'], hotpixels_log['right']['output'])

    fit_left = lanes.polynomial_fit(hotpixels_log['left'])
    fit_right = lanes.polynomial_fit(hotpixels_log['right'])
    #Get predicted lanes position as a function of height
    var_pts = np.linspace(0, warped_img.shape[0]-1, num=warped_img.shape[0]) #x-axis (height) points
    #predicted left lane
    pred_leftLane = fit_left['a2']*var_pts**2 + fit_left['a1']*var_pts + fit_left['a0']
    pred_rightLane = fit_right['a2']*var_pts**2 + fit_right['a1']*var_pts + fit_right['a0'] 
    
    if showMe:
        #display lane and best polynomial fits
        plt.imshow(warped_img, cmap='gray')
        plt.plot(pred_leftLane, var_pts, 'b-')
        plt.plot(pred_rightLane, var_pts, 'r-')
        plt.show()
    
    
    ###############
    # Draw line prediction/fit as thick lines
    h_pts = np.concatenate((var_pts, var_pts[::-1]), axis=0)
    left_lane = np.concatenate((pred_leftLane - draw_lane_thickness, pred_leftLane[::-1]+draw_lane_thickness), axis=0)
    l = np.array(list(zip(left_lane, h_pts)), np.int32)
    right_lane = np.concatenate((pred_rightLane - draw_lane_thickness, pred_rightLane[::-1]+draw_lane_thickness), axis=0)
    r = np.array(list(zip(right_lane, h_pts)), np.int32)
    road = np.zeros_like(input_image)
    cv2.fillPoly(road, [l], color=[255, 0, 0])
    cv2.fillPoly(road, [r], color=[255, 0, 0])
    img_size = (input_image.shape[1], input_image.shape[0])

    road_warped = cv2.warpPerspective(road, Minv, img_size, flags=cv2.INTER_LINEAR)
    result = cv2.addWeighted(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB), 0.9, road_warped, 1.0, 0.0)
    
    
    ###############
    # estimate offsetclose to driver (bottom of image)
    #print('***************************', pred_leftLane[-1])
    camera_center = (pred_leftLane[-1] +  pred_rightLane[-1])/2
    center_offset = (camera_center - warped_img.shape[1]/2) * xm_per_pix
    side_pos = 'left'
    if center_offset < 0:
        side_pos = 'right'
    
    cv2.putText(result, 'Vehicle is '+ str(round(center_offset, 3)) +'m ' +side_pos +'of center',
                (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255))
    
    ##############
    # Radius of Curvature
    # Determine polynomial parameters in real space
    real_space_pred_leftLane = lanes.polynomial_fit({'var': np.array(hotpixels_log['left']['var'])*xm_per_pix, 
                                                     'output': np.array(hotpixels_log['left']['output'])*ym_per_pix})
    
    real_space_pred_rightLane = lanes.polynomial_fit({'var': np.array(hotpixels_log['right']['var'])*xm_per_pix, 
                                                     'output': np.array(hotpixels_log['right']['output'])*ym_per_pix})
    
    # Calculate radii of curvature real space at point ref_point = bottom of image
    left_radius = ((1 + (2*real_space_pred_leftLane['a2']*ref_point_radius *ym_per_pix + real_space_pred_leftLane['a1'])**2)**1.5) \
                        / np.absolute(2*fit_left['a2'])
    
    right_radius = ((1 + (2*real_space_pred_rightLane['a2']*ref_point_radius*ym_per_pix + real_space_pred_rightLane['a1'])**2)**1.5) \
                        / np.absolute(2*real_space_pred_rightLane['a2'])
    
    middle_radius = (left_radius + right_radius)/2
    cv2.putText(result, 'Radius of curvature='+ str(round(middle_radius)) +'(m)', (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255))
    
       
    rad_past = history.update(middle_radius)

    
    if showMe:
        plt.title('Final Result')
        plt.imshow(result)
        plt.show()
    
    return result.astype('uint8')


def run_pipeline(input_image):
    input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
    ref_point_radius = input_image.shape[0]
    img_sz = (input_image.shape[1], input_image.shape[0])
    #input_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB)
    result = pipeline_predict_lanes(input_image, showMe=showMe)
    return result



inp = cv2.imread('project/YOURIMAGE0672.jpg')
out = pipeline_predict_lanes(inp)
plt.imshow(out)

print(rad_past)
```

```python
def MahalanobisDist(x, y):
    covariance_xy = np.cov(x,y, rowvar=0)
    inv_covariance_xy = np.linalg.inv(covariance_xy)
    xy_mean = np.mean(x),np.mean(y)
    x_diff = np.array([x_i - xy_mean[0] for x_i in x])
    y_diff = np.array([y_i - xy_mean[1] for y_i in y])
    diff_xy = np.transpose([x_diff, y_diff])
    
    md = []
    for i in range(len(diff_xy)):
        md.append(np.sqrt(np.dot(np.dot(np.transpose(diff_xy[i]),inv_covariance_xy),diff_xy[i])))
    return md

MahalanobisDist(l_var, l_out)

def MD_removeOutliers(x, y):
    MD = MahalanobisDist(x, y)
    threshold = np.mean(MD) * 1.9 # adjust 1.5 accordingly 
    nx, ny, outliers = [], [], []
    for i in range(len(MD)):
        if MD[i] <= threshold:
            nx.append(x[i])
            ny.append(y[i])
        else:
            outliers.append(i) # position of removed pair
    return (np.array(nx), np.array(ny), np.array(outliers))


a = MD_removeOutliers(r_var, r_out)

plt.plot(r_var, r_out, 'r.')
plt.plot(a[0], a[1], 'b.')

l_a, l_b, l_c = np.polyfit(r_var, r_out, 2)

x = np.linspace(0, 720, 719)
l_pred = l_a * x**2 + l_b * x + l_c
l_a1, l_b1, l_c1 = np.polyfit(a[0], a[1], 2)
l_pred1 = l_a1 * x**2 + l_b1 * x + l_c1
plt.plot(x, l_pred, 'y-')
plt.plot(x, l_pred1, 'k-')

#print(a[2])
```

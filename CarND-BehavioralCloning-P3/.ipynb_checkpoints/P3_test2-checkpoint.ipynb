```python


def set_angle_resolution(angle, precision):
    angle = float(angle)
    angle *= precision
    return int(angle) / precision

def plot_histogram_steering(angle_data, precision, bar_width=0.01):
    y = []
    for angle in angle_data:
        angle = set_angle_resolution(angle, precision)
        y.append(angle)
    # get unique values
    unique_angles = np.unique(y)
    # get histogram of y values
    hist = []
    for i in unique_angles :
        count = np.sum( y == i )
        hist.append([i, count])
    hist = np.array(hist)    
    #plot histogram
    _, ax = plt.subplots( 1, 2, figsize=(10, 5) )
    ax[0].set_title('Counts of steering angle occurence')
    ax[0].bar(hist[:,0], hist[:,1], width=bar_width, color='b')
    ax[1].bar(hist[:,0], np.log( hist[:,1] ), width=bar_width, color='b')
    ax[1].set_title('Log plot)')
    plt.show()



def angle_correction(original_steering, sign):
    return np.arctan( np.tan( original_steering ) + sign * 0.9/10)


data_path = 'data'

# open driving log file
#Gneerate data augmentation in log by including left/right cameras and corrected steering angle
with open(data_path + '/driving_log.csv', 'r') as csvfile:
    file_reader = csv.reader(csvfile, delimiter=',')
    driving_log = []
    
    #row = center_camera|left_camera|right_camera|steering|throttle|brake|speed
    for i, row in enumerate(file_reader):
        if i > 0: #skip header 
            center = [row[0], row[3], row[4]]  # file_name_center, steering angle
            driving_log.append(center)
            #left = [row[1].replace(" ", ""), angle_correction(angle, 1 )]  #file_name_left, corrected_steering
            #driving_log.append(left)
            #right = [row[2], angle_correction(angle, -1 )]  #file_name_right, corrected_steering
            #driving_log.append(right)

driving_log = np.array(driving_log)
np.save('log_short.npy', driving_log)

plot_histogram_steering( driving_log[:,1], precision=1000 )
plot_histogram_steering(driving_log[:,2], precision=1000)
# log file includes in order: center_camera|left_camera|right_camera|steering|throttle|brake|speed
# column of interest are center_camera and steering 
# create summary log with only img name and steering info - header removed
#driving_log = np.array( driving_log )

print('Dataset size: {}'.format(driving_log.shape ) )
# get column image name and reshape from (nbr_imgs,) to (nbr_imgs, 1)
# create a 2 column log with imgs/steering
```

```python
from keras.regularizers import l2
import os.path

def nvidia_base(img_sz, dropout = 0.8):
    '''
    Model based on Nvidia paper https://arxiv.org/pdf/1604.07316v1.pdf
    '''
    model = Sequential()
    model.add( Convolution2D(24, 5, 5, border_mode="valid", subsample=(2, 2), \
                             activation="elu", W_regularizer=l2(0.0001), input_shape=img_sz) )
    model.add( Convolution2D(36, 5, 5, border_mode="valid", subsample=(2, 2), \
                             W_regularizer=l2(0.0001), activation="elu") )
    model.add( Convolution2D(48, 5, 5, border_mode="valid", subsample=(2, 2), \
                             W_regularizer=l2(0.0001), activation="elu") )
    model.add( Convolution2D(64, 3, 3, border_mode="valid", subsample=(1, 1), \
                             W_regularizer=l2(0.0001), activation="elu") )
    model.add( Convolution2D(64, 3, 3, border_mode="valid", subsample=(1, 1), \
                             W_regularizer=l2(0.0001), activation="elu") )
    model.add(Flatten() )
    ## using dropout tends to make all-zero predictions! 
    model.add( Dense(1164, activation="elu", W_regularizer=l2(0.0001)) )
    Dropout(dropout)
    model.add( Dense(100, activation="elu", W_regularizer=l2(0.0001)) )
    Dropout(dropout)
    model.add( Dense(50, activation="elu", W_regularizer=l2(0.0001)) )
    Dropout(dropout)
    model.add( Dense(10, activation="elu", W_regularizer=l2(0.0001)) )
    Dropout(dropout)
    model.add( Dense(1, activation="linear", W_regularizer=l2(0.0001)) )
    model.compile(optimizer='adam', loss='mean_squared_error')

    print('Model NVidia-base is created and compiled..')
    return model



def continuousSteering(img_sz, dropout=1):
    # size of pooling area for max pooling
    pool_size = (2, 2)
    img_input = Input(shape = img_sz)
    x = Convolution2D(8, 5, 5, subsample=(1, 1), border_mode="valid", name='conv1', input_shape=img_sz, \
                           W_regularizer=l2(0.00001))(img_input)
    #model.add(Activation('elu'))
    x = Activation('relu')(x)
    #model.add(ZeroPadding2D((1, 1)))
    x = MaxPooling2D(pool_size=pool_size)(x)
    #model.add(Dropout(dropout))
    x = Convolution2D(8, 5, 5, subsample=(1, 1), border_mode="valid", W_regularizer=l2(0.00001))(x)
    #model.add(Activation('elu'))
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=pool_size)(x)
    #model.add(Dropout(dropout))
    x = Convolution2D(16, 4, 4, subsample=(1, 1), border_mode="valid",\
                           W_regularizer=l2(0.00001))(x)
    #model.add(Activation('elu'))
    x = Activation('relu')(x)
    x = MaxPooling2D(pool_size=pool_size)(x)
    #model.add(Dropout(dropout))
    x = Convolution2D(16, 5, 5, subsample=(1, 1), border_mode="valid",\
                            W_regularizer=l2(0.00001))(x)
    #model.add(Activation('elu'))
    x = Activation('relu')(x)
    
    x = Flatten()(x)
    x = Dropout(dropout)(x)
    
    x = Dense(128, W_regularizer=l2(0.0001))(x)
    #model.add(Activation('elu'))
    x = Activation('relu')(x)
    x = Dropout(dropout)(x)
    
    x = Dense(50, W_regularizer=l2(0.0001))(x)
    #model.add(Activation('elu'))
    x = Activation('relu')(x)
    x = Dropout(dropout)(x)
    
    x = Dense(10, W_regularizer=l2(0.0001))(x)
    #model.add(Activation('elu'))
    x = Activation('relu')(x)
    x = Dropout(dropout)(x)

    y_s = Dense(1, activation='linear', W_regularizer=l2(0.0001), init='he_normal', name='y_s')(x)
    
    x2 = Dense(300, W_regularizer=l2(0.0001))(x)
    #model.add(Activation('elu'))
    x2 = Activation('relu')(x2)
    x2 = Dropout(dropout)(x2)
    
    x2 = Dense(128, W_regularizer=l2(0.0001))(x2)
    #model.add(Activation('elu'))
    x2 = Activation('relu')(x2)
    x2 = Dropout(dropout)(x2)
    
    x2 = Dense(10, W_regularizer=l2(0.0001))(x2)
    #model.add(Activation('elu'))
    x2 = Activation('relu')(x2)
    x2 = Dropout(dropout)(x2)
    y_t = Dense(1, activation='linear', W_regularizer=l2(0.0001), init='he_normal', name='y_t')(x2)
   # optimizer = 
    model = Model(input=img_input, output=[y_s, y_t])
    model.compile(optimizer='adam', loss={'y_s': 'mean_squared_error', 'y_t': 'mean_squared_error'})
    
    print('Model -Home-Made- is created and compiled..')
    return model


def horizontal_flip(img, label):
    choice = np.random.choice([0,1])
    angle = label[0]
    if choice == 1:
        img, label[0] = cv2.flip(img, 1), -angle
    return (img, label)


def transf_brightness(img, label):
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
    #change Value/Brightness/Luminance: a*v + b
    alpha = np.random.uniform(low=0.1, high=1.0, size=None)
    v = hsv[:,:,2]
    v = v * alpha
    hsv[:,:,2] = v.astype('uint8')
    #min_val = np.min(hsv[:,:,2])
    #max_val = np.max(hsv[:,:,2])
    #print('min:{} | max: {}'.format(min_val, max_val))
    rgb = cv2.cvtColor(hsv.astype('uint8'), cv2.COLOR_HSV2RGB)
    return (rgb, label)

def center_RightLeft_swap(img_adress, label):
    # if 0(swap with left
    #1 swap with right
    # 2 no change
    swap = np.random.choice(['L', 'R', 'C'])
    angle = label[0]
    if swap == 'L':
        img_adress = img_adress.replace('center', 'left')
        label[0] = angle + 1.5/10
        return (img_adress, label)
    
    elif swap == 'R':
        img_adress = img_adress.replace('center', 'right')
        label[0] = angle - 1.5/10
        return (img_adress, label)
    
    else:
        return (img_adress, label)

def image_shift(img, label):
    # 
    #1 swap with right
    # 2 no change
    shift_direction = np.random.choice([-1, 0, 1])
    shift_amplitude = np.random.choice(np.arange(1, 11, 1))
    
    if shift_direction == -1:
        new_img = np.zeros(img.shape)
        new_img[shift_amplitude::, :, :] = img[shift_amplitude::, :, :]
        
    elif shift_direction == 1:
        new_img = np.zeros(img.shape)
        new_img[0:shift_amplitude, :, :] = img[0:shift_amplitude, :, :]

    else:
        return (img, label)


def filter_zero_steering(x, y, del_rate):
    # take away 80% of the zero steering angle images
    steering_zero_idx = np.where(y[:,0] == 0)
    steering_zero_idx = steering_zero_idx[0]
    #Randomly pick 80% of zero sterring angle images to delete
    size_del = int( len(steering_zero_idx) * del_rate )
    return np.random.choice(steering_zero_idx, size=size_del, replace=False)

    
driving_log = np.load( 'log_short.npy')
x = driving_log[:, 0]
y = driving_log[:, 1:3].astype(float)
#for i in range(len(y)):
#    ang = y[i]
#    y[i] = set_angle_resolution(ang, 1000)

# Generate Validation set

x, y = shuffle(x, y)
X_train, X_validation, y_train, y_validation = train_test_split(x, y, test_size=0.20, random_state=SEED)

print('Train set size: {} | Validation set size: {}'.format(len(X_train), len(X_validation)))
   
def batch_generator(x, y, batch_size, img_sz, data_dir='data/', monitor=True):
    """
    Yield X and Y data when the batch is filled.
    monitor: save last X and y batch 
    """
    y_bag = []
    x, y = shuffle(x, y)
    # take away 80% of the zero steering angle images
    rand_zero_idx = filter_zero_steering(x, y, del_rate=0.95)
    new_x = x
    new_y = y
    new_x = np.delete(new_x, rand_zero_idx, axis=0)
    new_y = np.delete(new_y, rand_zero_idx, axis=0)
    
    offset = 0
    while True:
        X = np.empty((batch_size, *img_sz))
        Y_s = np.empty((batch_size, 1))
        Y_t = np.empty((batch_size, 1))
        for i in range(batch_size):
            # random image swap: left-right-center
            img_adress, img_label = new_x[i + offset], new_y[i + offset]
            #assert os.path.exists(file_path), 'The file '+ img_adress +'not found - skip image.'
            img_adress, img_label = center_RightLeft_swap(img_adress, img_label )
            # Read img file
            img = cv2.imread(data_dir + img_adress)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            # change luminance: 50% chance
            img, img_label = transf_brightness(img, img_label)
            img, img_label = horizontal_flip(img, img_label)
            
            X[i,:,:,:] = cv2.resize(img[80:140, 0:320], (img_sz[0], img_sz[1]) ) / 255. - 0.5
            
            Y_s[i] = img_label[0]
            
            Y_t[i] = img_label[1]
            
            y_bag.append(img_label)
            if (i + 1) + offset > len(new_y) - 1:
                x, y = shuffle(x, y)
                
                rand_zero_idx = filter_zero_steering(x, y, del_rate=0.95)
                new_x = x
                new_y = y
                
                new_x = np.delete(new_x, rand_zero_idx, axis=0)
                new_y = np.delete(new_y, rand_zero_idx, axis=0)
                
                offset = 0
        yield (X, {'y_s': Y_s, 'y_t': Y_t})
        offset = offset + batch_size
        np.save('y_bag.npy', np.array(y_bag) )
        np.save('Xbatch_sample.npy', X ) #save last batch of images
            
            
def validation_gen(x, y, batch_size, data_dir='data/'):
    x, y = shuffle(x, y)
    offset = 0
    while True:
        X = np.empty((batch_size, *img_sz))
        if y != None:
            Y_s = np.empty((batch_size, 1))
            Y_t = np.empty((batch_size, 1))
        for i in range(batch_size):
            # random image swap: left-right-center
            img_adress, img_label = x[i + offset], y[i + offset]
            #assert os.path.exists(file_path), 'The file '+ img_adress +'not found - skip image.'
            img_adress, img_label = center_RightLeft_swap(img_adress, img_label )
            # Read img file
            img = cv2.imread(data_dir + img_adress)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            # change luminance: 50% chance
            img, img_label = transf_brightness(img, img_label)
            img, img_label = horizontal_flip(img, img_label)
            X[i,:,:,:] = cv2.resize(img[80:140, 0:320], (img_sz[0], img_sz[1]) ) / 255 - 0.5           
            Y_s[i] = img_label[0]
            Y_t[i] = img_label[1]
            if (i + 1) + offset > len(y) - 1:
                x, y = shuffle(x, y)
                offset = 0
        if y != None:
            yield (X, {'y_s': Y_s, 'y_t': Y_t} )
        else:
            yield X
        offset = offset + batch_size
    
              
########
# Parameters
#########
img_sz = (128, 128, 3)
batch_size = 128
samples_per_epoch = batch_size * 200
nb_val_samples = len(y_validation) - len(y_validation)%batch_size
nb_epoch = 10
print('samples per epoch', samples_per_epoch)

model = continuousSteering( img_sz )
print(model.summary())

# Callbacks: save best and early stop - based on validation loss value
model_path = os.path.expanduser('model.h5')
#Save the model after each epoch if the validation loss improved.
save_best = callbacks.ModelCheckpoint(model_path, monitor='val_loss', verbose=1, 
                                     save_best_only=True, mode='min')
#stop training if the validation loss doesn't improve for 5 consecutive epochs.
early_stop = callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=5, 
                                     verbose=0, mode='auto')
callbacks_list = [save_best, early_stop]



history = model.fit_generator(batch_generator(X_train, y_train, batch_size, img_sz),
                              samples_per_epoch=samples_per_epoch,
                              validation_data=validation_gen(X_validation, y_validation, batch_size=batch_size),
                              nb_val_samples=nb_val_samples,
                              nb_epoch=nb_epoch, verbose=1, callbacks=callbacks_list
                             )

#y_val_pred = model.predict_generator(validation_gen(X_validation, None, batch_size=batch_size), val_samples=nb_validation_samples)
#np.save('y_valid_pred.npy', y_val_pred)


architecture = model.to_json()
import json

with open('model.json', 'w') as f:
        f.write(architecture)
#save weight
model.save('model_.h5') 
print('Model (weight/architecture) saved!')
```

```python
from keras.models import Sequential
from keras.layers import Convolution2D, ZeroPadding2D, MaxPooling2D

img_width, img_height = 64, 64

# build the VGG16 network
model = Sequential()
model.add(ZeroPadding2D((1, 1), batch_input_shape=(1, img_height, img_width, 3)))
first_layer = model.layers[-1]
# this is a placeholder tensor that will contain our generated images
input_img = first_layer.input

# build the rest of the network
model.add(Convolution2D(64, 3, 3, activation='relu', name='conv1_1'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(64, 3, 3, activation='relu', name='conv1_2'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(128, 3, 3, activation='relu', name='conv2_1'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(128, 3, 3, activation='relu', name='conv2_2'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(256, 3, 3, activation='relu', name='conv3_1'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(256, 3, 3, activation='relu', name='conv3_2'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(256, 3, 3, activation='relu', name='conv3_3'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, 3, 3, activation='relu', name='conv4_1'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, 3, 3, activation='relu', name='conv4_2'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, 3, 3, activation='relu', name='conv4_3'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, 3, 3, activation='relu', name='conv5_1'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, 3, 3, activation='relu', name='conv5_2'))
model.add(ZeroPadding2D((1, 1)))
model.add(Convolution2D(512, 3, 3, activation='relu', name='conv5_3'))
model.add(MaxPooling2D((2, 2), strides=(2, 2)))

# get the symbolic outputs of each "key" layer (we gave them unique names).
layer_dict = dict([(layer.name, layer) for layer in model.layers])
```

```python
import h5py

weights_path = 'vgg16_weights.h5'

f = h5py.File(weights_path)
for k in range(f.attrs['nb_layers']):
    if k >= len(model.layers):
        # we don't look at the last (fully-connected) layers in the savefile
        break
    g = f['layer_{}'.format(k)]
    weights = [g['param_{}'.format(p)] for p in range(g.attrs['nb_params'])]
    model.layers[k].set_weights(weights)
f.close()
print('Model loaded.')
```

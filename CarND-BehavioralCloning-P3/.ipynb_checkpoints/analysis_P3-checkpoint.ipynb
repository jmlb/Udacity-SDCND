```python
def model_continuousSteering(img_sz):
    # size of pooling area for max pooling
    pool_size = (2, 2)
    print(img_sz)
    model = Sequential()

    model.add(Convolution2D(16, 8, 4, subsample=(4, 2), border_mode="same", name='conv1', input_shape=img_sz, activation='relu'))
    model.add(MaxPooling2D(pool_size=pool_size))

    model.add(Convolution2D(32, 5, 5, subsample=(2, 2), border_mode="same", activation='relu'))
    model.add(MaxPooling2D(pool_size=pool_size))

    model.add(Convolution2D(64, 5, 5, subsample=(2, 2), border_mode="same", activation='relu' ))
    #model.add(MaxPooling2D(pool_size=pool_size))

    model.add(Flatten())
    model.add(Dropout(.5))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(.5))

    model.add(Dense(1, activation='linear'))

    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

    print('Model relu2 is created and compiled..')
    return model


from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

driving_log_short = np.load( 'log_short.npy')


# Generate Validation set
SEED = 42

x, y = shuffle(x, y)
X_train, X_validation, y_train, y_validation = train_test_split(x, y, test_size=0.20, random_state=SEED)

def batch_generator(x, y, batch_size):
    """
    Yield X and Y data when the batch is filled.
    """
    while True:
        num_examples = len(y)
        x, y = shuffle(x, y)
        for offset in range(0, num_examples, batch_size):
            end = offset + batch_size
            X = np.zeros((batch_size, 160, 320, 3))
            Y = np.zeros((batch_size, 1))
        
            for i in range(batch_size):
                img = cv2.imread('data/'+x[i])
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                
                X[i,:,:,:] = img / 255. - 0.5
                Y[i] = y[i]
            yield (X, Y)

            
def validation_generator(x, y, batch_size):
    """
    Yield X and Y data when the batch is filled.
    """
    while True:
        num_examples = len(y)
        x, y = shuffle(x, y)
        for offset in range(0, num_examples, batch_size):
            end = offset + batch_size
            X = np.zeros((batch_size, 160, 320, 3))
            Y = np.zeros((batch_size, 1))
        
            for i in range(batch_size):
                img = cv2.imread('data/'+x[i])
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                
                X[i,:,:,:] = img / 255. - 0.5
                Y[i] = y[i]
            yield (X, Y)
            
model = model_continuousSteering( (160, 320, 3) )
print(model.summary())

print("loading images and labels")
batch_size = 128
samples_per_epoch = len(driving_log_short) - len(driving_log_short)%batch_size

history = model.fit_generator(batch_generator(X_train, y_train, batch_size), samples_per_epoch=samples_per_epoch, \
                              nb_epoch=15, verbose=1 )
#save weights
model.save_weights('model.h5')
#save model 
architecture = model.to_json()
import json

with open('model.json', 'w') as f:
        f.write(architecture)
```

# How to use the left and right camera for data augmentation

![GitHub Logo](Untitled1.png)

$$ tan(\alpha ') = tan( \alpha ) + \frac{\Delta x}{h} $$

We can assume an average car width of 1.8m and left and right camera at the extrem i.e at x=+/- 0.9m and the center camera at x=0.
The other assumpition is on h. We do not have any means to knwo that, so arbitrarily we will say that should correct within 5meters.

$$ tan(\alpha ') = tan( \alpha ) + \frac{0.9}{5} $$
$$ \alpha ' = arctan ( tan( \alpha ) + \frac{0.9}{5} ) =  $$

Below is an example:

```python
#original
original_angle = 0.2
correct_angle = np.arctan( np.tan( original_angle ) + 0.9/5)
print(correct_angle)
```

# How to do random shift for data augmentation

Using that data and our approximation on the distance left-center camera, we are going to calculate the shift in pixel: the shift center-left camera is $\Delta x = 70$ px.

$\frac{ \Delta x }{ h } = \frac{0.9}{5} = \frac{70}{389} = 0.18$


so:

$$\alpha '  = arctan( tan(\alpha) + \frac{\Delta x}{389} )$$

```python
from pathlib import Path
import cv2
import math
import cv2
from pathlib import Path
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

def set_angle_resolution(angle, precision):
    angle = float(angle)
    angle *= precision
    return int(angle) / precision

def plot_histogram_steering(angle_data, precision):
    y = []
    for angle in angle_data:
        angle = set_angle_resolution(angle, precision)
        y.append(angle)
    # get unique values
    unique_angles = np.unique(y)
    # get histogram of y values
    hist = []
    for i in unique_angles :
        count = np.sum( y == i )
        hist.append([i, count])
    hist = np.array(hist)    
    #plot histogram
    _, ax = plt.subplots( 1, 2, figsize=(10, 5) )
    ax[0].set_title('Counts of steering angle occurence')
    ax[0].bar(hist[:,0], hist[:,1], width=0.01, color='b')
    ax[1].bar(hist[:,0], np.log( hist[:,1] ), width=0.01, color='b')
    ax[1].set_title('Log plot)')
    plt.show()



def angle_correction(original_steering, sign):
    return np.arctan( np.tan( original_steering ) + sign * 0.9/10)


data_path = 'data'

# open driving log file
#Gneerate data augmentation in log by including left/right cameras and corrected steering angle
with open(data_path + '/driving_log.csv', 'r') as csvfile:
    file_reader = csv.reader(csvfile, delimiter=',')
    driving_log = []
    #row = center_camera|left_camera|right_camera|steering|throttle|brake|speed
    for i, row in enumerate(file_reader):
        if i > 0: #skip header
            angle = set_angle_resolution(row[3], 1000)
            center = [row[0], angle]  # file_name_center, steering angle
            driving_log.append(center)
            left = [row[1], angle_correction(angle, 1 )]  #file_name_left, corrected_steering
            driving_log.append(left)
            right = [row[2], angle_correction(angle, -1 )]  #file_name_right, corrected_steering
            driving_log.append(right)


driving_log = np.array(driving_log)
np.save('log_short.npy', driving_log)

plot_histogram_steering( driving_log[:,1], precision=1000 )

# log file includes in order: center_camera|left_camera|right_camera|steering|throttle|brake|speed
# column of interest are center_camera and steering 
# create summary log with only img name and steering info - header removed
#driving_log = np.array( driving_log )

print('Dataset size: {}'.format(driving_log.shape ) )
# get column image name and reshape from (nbr_imgs,) to (nbr_imgs, 1)
# create a 2 column log with imgs/steering
```

```python
from keras.regularizers import l2, activity_l2

def model_continuousSteering(img_sz):
    # size of pooling area for max pooling
    pool_size = (2, 2)
    print(img_sz)
    model = Sequential() 
    model.add(Convolution2D(16, 5, 5, subsample=(1, 1), border_mode="same", name='conv1', input_shape=img_sz, \
                           W_regularizer=l2(0.001)))
    model.add(PReLU())
    model.add(MaxPooling2D(pool_size=pool_size))
    model.add(Dropout(.5))
    model.add(Convolution2D(32, 5, 5, subsample=(1, 1), border_mode="same", W_regularizer=l2(0.001)))
    model.add(PReLU())
    model.add(MaxPooling2D(pool_size=pool_size))
    model.add(Dropout(.5))
    model.add(Convolution2D(64, 5, 5, subsample=(1, 1), border_mode="same", W_regularizer=l2(0.001)))
    model.add(PReLU())
    #model.add(MaxPooling2D(pool_size=pool_size))

    model.add(Flatten())
    model.add(Dropout(.5))
    model.add(Dense(256, W_regularizer=l2(0.001)))
    model.add(PReLU())
    model.add(Dropout(.5))

    model.add(Dense(1, activation='linear'))

    model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

    print('Model relu2 is created and compiled..')
    return model



driving_log = np.load( 'log_short.npy')
x = driving_log[:, 0]
y = driving_log[:, 1].astype(float)

for i in range(len(y)):
    ang = y[i]
    y[i] = set_angle_resolution(ang, 1000)

# Generate Validation set

x, y = shuffle(x, y)
X_train, X_validation, y_train, y_validation = train_test_split(x, y, test_size=0.20, random_state=SEED)

print('Train set size: {} | Validation set size: {}'.format(len(X_train), len(X_validation)))
print(x)
def batch_generator(x, y, batch_size):
    """
    Yield X and Y data when the batch is filled.
    """
    elt0 = np.where(y == 0)
    elt0 = elt0[0]
    while True:
        num_examples = len(y)
        rand_elt0 = np.random.choice(elt0, size=(len(elt0) - 250), replace=False)
        new_x = x
        new_y = y
        #new_x = np.delete(new_x, rand_elt0, axis=0)
        #new_y = np.delete(new_y, rand_elt0, axis=0)
        new_x, new_y = shuffle(x, y)
        for offset in range(0, num_examples, batch_size):
            end = offset + batch_size
            X = np.zeros((batch_size, 64, 64, 3))
            Y = np.zeros((batch_size, 1))
        
            for i in range(batch_size):
                img = cv2.imread('data/'+new_x[i].replace(" ", ""))
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                
                X[i,:,:,:] = cv2.resize(img[50:130, 0:320], (64, 64) )              
                #X[i,:,:,:] = img
                Y[i] = new_y[i]
            X = X / 127.5 - 1
            yield (X, Y)

def validation_gen(x, y, batch_size):
    while True:
        num_examples = len(x)
        for offset in range(0, num_examples, batch_size):
            end = offset + batch_size
            X = np.zeros((batch_size, 64, 64, 3))
            # for evaluation / for prediction y is None
            if y != None:
                Y = np.zeros((batch_size, 1))
        
            for i in range(batch_size):
                img = cv2.imread('data/'+x[i].replace(" ", ""))
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                #X[i,:,:,:] = cv2.resize( img[50:130, 0:320], None,(64, 64), interpolation = cv2.INTER_CUBIC)
                X[i,:,:,:] = cv2.resize( img[50:130, 0:320], (64, 64) )
                #X[i,:,:,:] = img
                if y != None:
                    Y[i] = y[i]
            X = X.astype('float32')
            X = X / 127.5 - 1
            if y != None:
                yield (X, Y)
            else:
                yield X
            
model = model_continuousSteering( (64, 64, 3) )
print(model.summary())
batch_size = 128
print("loading images and labels")
samples_per_epoch = len(X_train) - len(X_train)%batch_size
print(samples_per_epoch)
history = model.fit_generator(batch_generator(X_train, y_train, batch_size), samples_per_epoch=samples_per_epoch, \
                              nb_epoch=5, verbose=1, validation_data=validation_gen(X_validation, y_validation, batch_size=batch_size), 
                            nb_val_samples=len(y_validation))

nb_validation_samples = len(X_validation) - len(X_validation)%batch_size
y_val_pred = model.predict_generator(validation_gen(X_validation, None, batch_size=batch_size), val_samples=nb_validation_samples)
np.save('y_valid_pred.npy', y_val_pred)
#save weights
model.save_weights('model.h5')
#save model 
architecture = model.to_json()
import json

with open('model.json', 'w') as f:
        f.write(architecture)
```
